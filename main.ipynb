```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer,ENGLISH_STOP_WORDS
import re
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv("C:/ML_Dataset/food_review_dataset.csv")

corpus=df.Review
target=df.Liked

myspw=list(ENGLISH_STOP_WORDS)
myspw.remove('not')

def mycleaning(doc):
    return re.sub("[^a-zA-Z ]","",doc).lower()

cv=CountVectorizer(stop_words=myspw,preprocessor=mycleaning)
model=LogisticRegression()

pipe=Pipeline([('vectorizer',cv),('model',model)])
pipe.fit(corpus,target)

sample=input("enter review")
pred=pipe.predict([sample])
prob=pipe.predict_proba([sample])
if pred[0]==0:
    print("Neg")
    print(f"Confidence Score : {prob[0][0]:.2f}")
else:
    print("Pos") 
    print(f"Confidence Score : {prob[0][1]:.2f}")
```

```python
cross_val_score(pipe,corpus,target,cv=5).mean()
```
